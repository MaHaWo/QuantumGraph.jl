# Configuration files

Load QuantumGraph configuration tags from actual YAML files under `docs/examples/configs/`.


```julia
function find_repo_root(start = pwd())
    dir = abspath(start)
    while !isfile(joinpath(dir, "Project.toml")) || !isdir(joinpath(dir, "src"))
        parent = dirname(dir)
        parent == dir && error("Could not find QuantumGraph.jl repository root")
        dir = parent
    end
    dir
end

repo_root = find_repo_root()
cd(repo_root)
using Pkg
Pkg.activate(repo_root)
```

```julia
using QuantumGraph

config_path = joinpath(repo_root, "docs", "src", "examples", "configs", "basic_tags.yml")
config = load_config(read(config_path, String))
config
```

```julia
expanded = expand_config(config)
length(expanded), first(expanded)
```

```julia
resolved = resolve_config(Dict("width" => 32, "copied_width" => Reference(["width"])))
resolved
```

```julia
struct IdentityActivation end
QuantumGraph.register_object!("Docs.IdentityActivation", IdentityActivation())
resolve_config(Dict("activation" => ObjectReference("Docs.IdentityActivation")))
```
